```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# xpoints = np.array([5,7,10])
ypoints = np.array([0,7,3,8,4,5,6])

plt.plot(ypoints,"ok", ms =20, mec="r",ls=":")
plt.show()
```

```python
x1=np.array([10,13,15,17])
y1=np.array([5,7,8,11])
x2=np.array([4,16,18,14])
y2=np.array([4,9,12,19])

plt.subplot(2,2,1)

plt.plot(x1,y1,marker="o")
plt.title("Line Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(c="g", lw="1" , ls=":")

plt.subplot(2,2,2)
plt.plot(x2,y2,marker="o")
plt.title("Line Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(c="g", lw="1" , ls=":")
plt.subplot(2,2,3)
plt.plot(x2,y2,marker="o")
plt.title("Line Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(c="g", lw="1" , ls=":")
plt.subplot(2,2,4)
plt.plot(x2,y2,marker="o")
plt.title("Line Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(c="g", lw="1" , ls=":")
plt.show()
```

```python
x=np.array(range(0,10))
y=np.array([1,23,12,25,19,16,7,10,12,18])
color = np.array([8,6,10,9,4,8,3,7,1,2])
size = np.array([8,6,10,9,4,8,3,7,1,2])*20

plt.scatter(x,y,c=color, s= size,cmap="nipy_spectral")
plt.colorbar()
plt.show()
```
