```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
data
```

```python
data.info()
data.isna().sum()
```

```python
print(data.duplicated().sum())
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
sns.lineplot(data=data,x="total_bill",y="tip")
plt.title("TIPS VS BILL")
plt.show()
```

```python
sns.scatterplot(data=data,x="tip",y="total_bill",hue="time")
plt.show()
```

```python
sns.barplot(data,x="day",y="total_bill")
plt.show()
```

```python
sns.countplot(data,x="day",hue="time")
plt.show()
```

```python
data[data["time"] == "Dinner"]["total_bill"].mean()
```

```python
sns.histplot(data,x="total_bill",kde=True)
plt.show()
```

```python
sns.kdeplot(data,x="total_bill", fill=True)
plt.show()
```

```python
sns.boxplot(data,x="day",y="total_bill")
plt.show()
```

```python
import numpy as np
arr = np.array([1,7,7,8,9,10,11,11,12,20])

sns.boxplot(arr)
plt.show()

Q1 = np.quantile(arr,0.25)
Q3 = np.quantile(arr,0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5*IQR
lower = Q1 - 1.5*IQR
print(upper)
print(lower)

# list = np.where(arr <= upper)
arr = arr[arr <= upper]
arr = arr[arr >= lower]

print(arr)
```

```python

sns.boxplot(arr)
plt.show()
```
