```python
df["Date"] = pd.to_datetime(df["Date"],format = "mixed")
mean = df['Calories'].mean()
df.fillna({"Calories":mean},inplace=True)
df.dropna(subset=["Date"],inplace=True)
df.info()
df
```

```python
mod = df["Duration"].mode()[0]
df.loc[7,"Duration"] = mod

for i in df.index:
    dur = df.loc[i,"Duration"]
    if dur > 120:
        df.loc[i,"Duration"]=mod
df
```
